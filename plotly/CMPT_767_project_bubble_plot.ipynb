```python
import pandas as pd
import plotly.express as px
from google.colab import drive
import os
import numpy as np
```

```python
drive.mount('/content/drive')
BASE_DIR = '/content/drive/My Drive/CMPT_767/data'

filename = os.path.join(BASE_DIR, "transform_df.csv")
print(filename)
```

### data clean

```python
df_origin = pd.read_csv(filename)
display(df_origin)
print(df_origin.columns)
```

```python
df = df_origin[['DAY','MONTH','YEAR', 'STATION', 'STATIONNAME', 'STATE','MonthlyAvgTemp']]
display(df)
```

```python
df_final = df.groupby(by=['DAY','MONTH',"YEAR",'STATE']).mean()
if "STATION" in df_final.columns:
  df_final = df_final.drop("STATION",axis=1)

df_final = df_final.reset_index()
display(df_final)
```

### set color

```python
color_setting_df = df_final[["MonthlyAvgTemp"]].copy()
color_setting_df = color_setting_df.describe()
color_scale = color_setting_df.iloc[3:8]
display(color_scale)
color_scale_list = color_scale['MonthlyAvgTemp'].to_list()
```

### plot graph

```python
display(df_final)
```

```python
df_final['MONTH'] = df_final['MONTH'].astype(str)
df_final['YEAR'] = df_final['YEAR'].astype(str)
df_final['DAY'] = df_final['DAY'].astype(str)
df_final["DATE"] = pd.concat([df_final["DAY"], df_final["MONTH"], df_final["YEAR"]], axis=1).apply(lambda row: '/'.join(row), axis=1)
df_final["DATE"] = pd.to_datetime(df_final["DATE"], format="%d/%m/%Y")
display(df_final)
```

```python
df_final['temperature_bin'] = pd.cut(df_final["MonthlyAvgTemp"], bins=color_scale_list, labels=['Frigid','Cool','Warm','Scorching'])
display(df_final)
```

```python
data = {'STATE': df_final["STATE"].to_list(),
        'DATE': df_final["DATE"].to_list(),
        'Avg_Temp':df_final["MonthlyAvgTemp"].to_list(),
        'Avg_abs_Temp': df_final["MonthlyAvgTemp"].abs().to_list(),
        'Category':df_final["temperature_bin"].to_list()}

df_to_chart = pd.DataFrame(data)
df_to_chart = df_to_chart.sort_values(by=['DATE', 'STATE'])
df_to_chart['format_date'] = df_to_chart['DATE'].apply(lambda row: row.strftime("%Y-%m"))
display(df_to_chart)
```

```python


# Define a color map for categories
color_map = {'Frigid': '#083d6d', "Cool": '#83ceee', "Warm": '#fb9c4c',"Scorching": '#e31e26'}

# Create a scatter plot with custom colors for each category
fig = px.scatter(df_to_chart, x='STATE', y='format_date', color='Category', size='Avg_abs_Temp',
                 title='Bubble Plot with Average Monthly Temperature in USA', hover_name="STATE", hover_data={"DATE":True,
                 "Avg_Temp":True, 'Avg_abs_Temp':False, 'STATE':False, 'Category':False,'format_date':False}, width=1160, height=1000,
                 color_discrete_map=color_map)
fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = df_to_chart["STATE"].to_list()
    ),
    yaxis = dict(
        tickmode = 'array',
        tickvals = df_to_chart["format_date"].to_list(),
        ticktext = df_to_chart["format_date"].astype(str)
    )
)

# Show the plot
fig.show()
fig.write_html("bubble_plot.html")
```

### Another set

```python
df = df_origin[['DAY','MONTH','YEAR', 'STATION', 'STATIONNAME', 'STATE','MonthlyAvgPreciptation']]
display(df)
```

```python
df_final = df.groupby(by=['DAY','MONTH',"YEAR",'STATE']).mean()
if "STATION" in df_final.columns:
  df_final = df_final.drop("STATION",axis=1)

df_final = df_final.reset_index()
display(df_final)
```

```python
color_setting_df = df_final[["MonthlyAvgPreciptation"]].copy()
color_setting_df = color_setting_df.describe()
color_scale = color_setting_df.iloc[3:8]
display(color_scale)
color_scale_list = color_scale['MonthlyAvgPreciptation'].to_list()
print(color_scale_list)
```

```python
display(df_final)
```

```python
df_final['MONTH'] = df_final['MONTH'].astype(str)
df_final['YEAR'] = df_final['YEAR'].astype(str)
df_final['DAY'] = df_final['DAY'].astype(str)
df_final["DATE"] = pd.concat([df_final["DAY"], df_final["MONTH"], df_final["YEAR"]], axis=1).apply(lambda row: '/'.join(row), axis=1)
df_final["DATE"] = pd.to_datetime(df_final["DATE"], format="%d/%m/%Y")
display(df_final)
```

```python
df_final['precip_bin'] = pd.cut(df_final["MonthlyAvgPreciptation"], bins=color_scale_list, labels=['Drizzle','Shower','Rain','Downpour'])
display(df_final)
```

```python
data = {'STATE': df_final["STATE"].to_list(),
        'DATE': df_final["DATE"].to_list(),
        'Avg_Precip':df_final["MonthlyAvgPreciptation"].to_list(),
        'Category':df_final["precip_bin"].to_list()}

df_to_chart = pd.DataFrame(data)
df_to_chart = df_to_chart.sort_values(by=['DATE', 'STATE'])
df_to_chart['format_date'] = df_to_chart['DATE'].apply(lambda row: row.strftime("%Y-%m"))
display(df_to_chart)
```

```python


# Define a color map for categories
color_map = {'Drizzle': '#9dc6d8', "Shower": '#8bd8fb', "Rain": '#03a9f5',"Downpour": '#2196f3'}

# Create a scatter plot with custom colors for each category
fig = px.scatter(df_to_chart, x='STATE', y='format_date', color='Category', size='Avg_Precip',
                 title='Bubble Plot with Average Preciptation in USA', hover_name="STATE", hover_data={"DATE":True,
                 "Avg_Precip":True, 'STATE':False, 'Category':False, 'format_date':False}, width=1160, height=1000,
                 color_discrete_map=color_map)
fig.update_layout(
    yaxis = dict(
        tickmode = 'array',
        tickvals = df_to_chart["format_date"].to_list(),
        ticktext = df_to_chart["format_date"].astype(str)
    )
)

# Show the plot
fig.show()
fig.write_html("bubble_plot.html")
```
